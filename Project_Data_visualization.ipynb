# **Data loading and inspection using Pandas**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Ardi1623/project-data-visualization/refs/heads/main/marketing_campaign.csv", sep="\t")
```

# **Handling missing values**

```python
# Check for null values in each column
null_counts = data.isnull().sum()

# Display the results
print(null_counts)
```

```python
#separate Null Income
df_customers_missing_income = data[data.Income.isna()]
df_customers_clean_income = data[data.Income.notna()]
```

# **Data cleaning and transformation**

```python
# Check for duplicate rows
duplicates_MI = df_customers_missing_income.duplicated().sum()
duplicates_CI = df_customers_clean_income.duplicated().sum()
print(f"Number of duplicate rows: {duplicates_MI}")
print(f"Number of duplicate rows: {duplicates_CI}")

# Drop duplicate rows if any
df_customers_missing_income = df_customers_missing_income.drop_duplicates()
df_customers_clean_income = df_customers_clean_income.drop_duplicates()
df_customers_missing_income.info()
df_customers_clean_income.info()
```

```python
# Remove rows with invalid years
df_customers_missing_income = df_customers_missing_income[(df_customers_missing_income['Year_Birth'] >= 1900) & (df_customers_missing_income['Year_Birth'] <= 2023)]
df_customers_clean_income = df_customers_clean_income[(df_customers_clean_income['Year_Birth'] >= 1900) & (df_customers_clean_income['Year_Birth'] <= 2023)]

```

# **Exploratory data analysis**

**Define at least two relevant business questions based on the dataset.**

I want to identify spending based on generation **year_birth**, **education** and **marital_status**.

Therefore, i need to identify generation based on **year_birth**

```python
# Define a function to map Year_Birth to a generation
def map_generation(year):
    if 1928 <= year <= 1945:
        return "Silent Generation"
    elif 1946 <= year <= 1964:
        return "Baby Boomers"
    elif 1965 <= year <= 1980:
        return "Generation X"
    elif 1981 <= year <= 1996:
        return "Millennials"
    elif 1997 <= year <= 2012:
        return "Generation Z"
    else:
        return "Unknown"

# Apply the function to create a new column
df_customers_missing_income['Generation'] = df_customers_missing_income['Year_Birth'].apply(map_generation)
df_customers_clean_income['Generation'] = df_customers_clean_income['Year_Birth'].apply(map_generation)
```

*From now on, we continue with data without 0/NULL income*

After we group the data **Generation** we need to categorize the spending by Low, Medium, High, Very High

```python
# Categorize MntMeatProducts by quantile
df_customers_clean_income['Wines_Spending_Category'] = pd.qcut(
    df_customers_clean_income['MntWines'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High']
)
df_customers_clean_income['Fruit_Spending_Category'] = pd.qcut(
    df_customers_clean_income['MntFruits'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High']
)
df_customers_clean_income['Meat_Spending_Category'] = pd.qcut(
    df_customers_clean_income['MntMeatProducts'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High']
)
df_customers_clean_income['Fish_Spending_Category'] = pd.qcut(
    df_customers_clean_income['MntFishProducts'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High']
)
df_customers_clean_income['Sweet_Spending_Category'] = pd.qcut(
    df_customers_clean_income['MntSweetProducts'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High']
)
df_customers_clean_income['Gold_Spending_Category'] = pd.qcut(
    df_customers_clean_income['MntGoldProds'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High']
)
```

We will focus on low spending to identify areas for improvement.

```python
# Count the number of 'Low' in each spending category
Low_counts = {
    'Wines_Spending_Category': df_customers_clean_income['Wines_Spending_Category'].value_counts().get('Low', 0),
    'Fruit_Spending_Category': df_customers_clean_income['Fruit_Spending_Category'].value_counts().get('Low', 0),
    'Meat_Spending_Category': df_customers_clean_income['Meat_Spending_Category'].value_counts().get('Low', 0),
    'Fish_Spending_Category': df_customers_clean_income['Fish_Spending_Category'].value_counts().get('Low', 0),
    'Sweet_Spending_Category': df_customers_clean_income['Sweet_Spending_Category'].value_counts().get('Low', 0),
    'Gold_Spending_Category': df_customers_clean_income['Gold_Spending_Category'].value_counts().get('Low', 0)
}

# Identify the category with the most 'Low'
max_category = min(Low_counts, key=Low_counts.get)
max_count = Low_counts[max_category]

# Create a bar plot to visualize the counts of 'Low' in each category
plt.figure(figsize=(10, 6))
sns.barplot(x=list(Low_counts.keys()), y=list(Low_counts.values()), palette='viridis')

# Title and labels
plt.title(f"Spending Category with Most 'Low' (Category: {max_category})", fontsize=16)
plt.xlabel("Spending Categories", fontsize=12)
plt.ylabel("Count of 'Low'", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()

# Print which category has the most 'Low' values
print(f"The category with the most 'Low' is: {max_category} with {max_count} entries.")
```

Based on the diagram above, customers appear to spend the least on Fruit.

Now We Deepening the analyze to people at home

```python
# Add a new column 'People_at_home' as the sum of Kidhome and Teenhome plus 2
df_customers_clean_income['People_at_home'] = df_customers_clean_income['Kidhome'] + df_customers_clean_income['Teenhome'] + 2
```

we continue the analyze from marital_status and people_at_home population compare to the lowest purchase

```python
# Filter for rows where 'Wines_Spending_Category' is 'Low'
df_low_spending = df_customers_clean_income[df_customers_clean_income['Wines_Spending_Category'] == 'Low']

# Group by Marital_Status, People_at_home, and Wines_Spending_Category and count the occurrences
grouped_data = df_low_spending.groupby(['Marital_Status', 'People_at_home', 'Wines_Spending_Category']).size().reset_index(name='Count')

# Visualization of the lowest spending category by Marital_Status and People_at_home
plt.figure(figsize=(12, 6))
sns.barplot(x='Marital_Status', y='Count', hue='People_at_home', data=grouped_data, palette='coolwarm')
plt.title('Lowest Spending Category by Marital Status and People at Home', fontsize=16)
plt.xlabel('Marital Status', fontsize=12)
plt.ylabel('Count of Low Spending (Wines)', fontsize=12)
plt.legend(title="People at Home", loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print a summary of the lowest spending category (Wines_Spending_Category = 'Low')
lowest_spending_summary = grouped_data[['Marital_Status', 'People_at_home', 'Count']].sort_values(by='Count', ascending=False)
print(lowest_spending_summary)
```

Based on above output
we will maximize the spending on the highest population
Married_Status = Married and people_at_home=3

so we need to identify what category they most spend on

```python
# Filter for married individuals with 3 people at home
df_filtered = df_customers_clean_income[(df_customers_clean_income['Marital_Status'] == 'Married') & (df_customers_clean_income['People_at_home'] == 3)]

# Count the number of 'Very High' in each spending category
very_high_counts = {
    'Wines_Spending_Category': df_filtered['Wines_Spending_Category'].value_counts().get('Very High', 0),
    'Fruit_Spending_Category': df_filtered['Fruit_Spending_Category'].value_counts().get('Very High', 0),
    'Meat_Spending_Category': df_filtered['Meat_Spending_Category'].value_counts().get('Very High', 0),
    'Fish_Spending_Category': df_filtered['Fish_Spending_Category'].value_counts().get('Very High', 0),
    'Sweet_Spending_Category': df_filtered['Sweet_Spending_Category'].value_counts().get('Very High', 0),
    'Gold_Spending_Category': df_filtered['Gold_Spending_Category'].value_counts().get('Very High', 0)
}

# Identify the category with the highest 'Very High' count
max_category = max(very_high_counts, key=very_high_counts.get)
max_count = very_high_counts[max_category]

# Visualize the counts of 'Very High' in each category
plt.figure(figsize=(10, 6))
sns.barplot(x=list(very_high_counts.keys()), y=list(very_high_counts.values()), palette='viridis')

# Title and labels
plt.title(f"Spending Category with Most 'Very High' for Married Individuals with 3 People at Home (Category: {max_category})", fontsize=16)
plt.xlabel("Spending Categories", fontsize=12)
plt.ylabel("Count of 'Very High'", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()

# Print the category with the most 'Very High' spending
print(f"The category with the most 'Very High' spending is: {max_category} with {max_count} entries.")
```

From above data, we found that Married with 3 People mostly spend their income to gold

now we identify which sales channel they buy at most

```python
# Filter for married individuals with 3 people at home
df_filtered = df_customers_clean_income[(df_customers_clean_income['Marital_Status'] == 'Married') & (df_customers_clean_income['People_at_home'] == 3)]

# Columns to check for purchase types
purchase_columns = ['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']

# Calculate the frequency of each purchase type
purchase_frequencies = df_filtered[purchase_columns].apply(pd.Series.value_counts).fillna(0).T

# Get the most frequent purchase type for each category
purchase_frequencies['Most_Frequent_Purchase'] = purchase_frequencies.idxmax(axis=1)
purchase_frequencies['Frequency'] = purchase_frequencies.max(axis=1)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=purchase_frequencies.index, y=purchase_frequencies['Frequency'], palette='coolwarm')
plt.title("Most Frequent Purchase Types for Married Individuals with 3 People at Home", fontsize=16)
plt.xlabel("Purchase Type", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()

# Print the results of the most frequent purchase types
print(purchase_frequencies[['Most_Frequent_Purchase', 'Frequency']])
```
